```python
import tensorflow as tf
import matplotlib.pyplot as plt
from PIL import Image

# Load the generator model
model = tf.keras.models.load_model('SoftconExpoModel.h5')

# Load and preprocess the test input image
input_image = tf.io.read_file('./data/val/test.png')
input_image = tf.image.decode_png(input_image)
input_image = input_image[..., :3]  # Keep only the first 3 channels
input_image = tf.image.resize(input_image, [256, 256])
input_image = tf.cast(input_image, tf.float32)
input_image = (input_image / 127.5) - 1
input_image = tf.expand_dims(input_image, 0)  # Add batch dimension
  # Add batch dimension

# Generate the predicted image
predicted_image = model(input_image, training=True)[0]

# Display the predicted image
plt.figure(figsize=(5, 5))
plt.imshow(predicted_image * 0.5 + 0.5)  # Display the predicted image
plt.title('Predicted Image')
plt.axis('off')
plt.show()
```
